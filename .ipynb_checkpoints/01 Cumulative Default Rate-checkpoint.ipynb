## Side Project: Cumulative Default Rate
### Import & Clean data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

import seaborn as sns

%matplotlib inline

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 10)
```

```python
loan11 = pd.read_csv('LOAN/loan11.csv', header = 1, skipfooter = 2)
loan13 = pd.read_csv('LOAN/loan13.csv', header = 1, skipfooter = 2)
loan14 = pd.read_csv('LOAN/loan14.csv', header = 1, skipfooter = 2)
loan15 = pd.read_csv('LOAN/loan15.csv', header = 1, skipfooter = 2)
loan16q1 = pd.read_csv('LOAN/loan16q1.csv', header = 1, skipfooter = 2)
loan16q2 = pd.read_csv('LOAN/loan16q2.csv', header = 1, skipfooter = 2)
loan16q3 = pd.read_csv('LOAN/loan16q3.csv', header = 1, skipfooter = 2)
loan16q4 = pd.read_csv('LOAN/loan16q4.csv', header = 1, skipfooter = 2)
loan17q1 = pd.read_csv('LOAN/loan17q1.csv', header = 1, skipfooter = 2)
```

```python
Series = [loan11, loan13, loan14, loan15, loan16q1, loan16q2, loan16q3, loan16q4, loan17q1]
fullData = pd.concat(Series, ignore_index = True)
```

```python
cols_to_keep = ["loan_status", "last_pymnt_d", "issue_d", "term", "grade"]
df = fullData[cols_to_keep]
df = df[df['loan_status'].isin(["Fully Paid", "Charged Off"])].copy()
```

```python
TimeColumn = pd.Series(['issue_d', 'last_pymnt_d'])
df[TimeColumn] = df[TimeColumn].apply(pd.to_datetime)
df["issue_d"] = pd.PeriodIndex(df["issue_d"], freq='M')
df["last_pymnt_d"] = pd.PeriodIndex(df["last_pymnt_d"], freq='M')
df["loan_month"] = df["last_pymnt_d"] - df["issue_d"]
```

```python
df = df.dropna(axis = 0)
```

### Calculate Cumulative Default Rate

```python
loan = df.query("loan_status != 'Current'")[['term', 'loan_status', 'loan_month', 'grade']]
loan_df = loan.pivot_table(index = ['term', 'grade', 'loan_month'], columns=['loan_status'], aggfunc=len, fill_value=0, dropna = False).reset_index()
loan_df['Charged Off_cumsum'] = loan_df.groupby(['term', 'grade'])['Charged Off'].transform(pd.Series.cumsum)
loan_cnt = loan.groupby(['term', 'grade'],  as_index = False)["loan_status"].agg(len).rename(columns = {"loan_status": "loan_counts"})
loan_df = pd.merge(loan_df, loan_cnt, on = ['term', 'grade'])
loan_df['Default Rate'] = loan_df['Charged Off_cumsum'] / loan_df['loan_counts']
loan_df = loan_df.drop(["Charged Off", "Charged Off_cumsum", "Fully Paid", "loan_counts"], 1)
```

### Plots

```python
df36 = loan_df[loan_df["term"] == " 36 months"]
plt.figure(figsize=(30,20))
sns.pointplot(x = "loan_month", y = "Default Rate", hue = "grade",
              data = df36, fit_reg = False, join = True, ci = None,
              palette = "YlGnBu_d").set_title("36 Months", size = 40)

plt.plot([36, 36], [0, 0.5], linewidth = 2, color = "gray", linestyle='dashed')
plt.xlabel('# of Month', fontsize = 30)
plt.ylabel('Cumulative Default Rate', fontsize = 30)
```

```python
df60 = loan_df[loan_df["term"] == " 60 months"]
plt.figure(figsize=(30,20))
sns.pointplot(x = "loan_month", y = "Default Rate", hue = "grade",
              data = df60, fit_reg = False, join = True, ci = None,
              palette = "YlGnBu_d").set_title("60 Months", size = 40)

plt.plot([60, 60], [0, 0.5], linewidth = 2, color = "gray", linestyle='dashed')
plt.xlabel('# of Month', fontsize = 30)
plt.ylabel('Cumulative Default Rate', fontsize = 30)
```
